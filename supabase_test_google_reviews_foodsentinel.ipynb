```python
%%writefile run_models.py
# PASTE THE FULL UPDATED CODE HERE
#!/usr/bin/env python3
"""
run_models.py (Colab-friendly, public Storage, skip missing models)

Env vars required:
  SUPABASE_URL="https://<project>.supabase.co"
  SUPABASE_STORAGE_URL="https://<project>.supabase.co/storage/v1"   (recommended)
  SUPABASE_DB_URL="postgresql://...?...sslmode=require"
  SUPABASE_SERVICE_ROLE_KEY="..."  (recommended if DB has RLS / restricted perms)

Storage downloads use:
  /storage/v1/object/public/<bucket>/<path>  [web:444]
"""

import os
import sys
import io
import time
import logging
from typing import List, Tuple, Dict, Any

import requests
import joblib
import numpy as np
import psycopg2
from psycopg2 import sql
from psycopg2.extras import execute_values

# =======================
# Config
# =======================
BUCKET_NAME = "safefoodsentinel_bucket_models"
MODEL_KEYS = {
    "sentiment": "1.sentiment_model_lovable_tfidf.pkl",
    "food_safety": "2.food_safety_model_ovr_logreg_lovable.pkl",
    "svc_foodqual": "3.service_foodquality_multioutput_model_lovable.pkl",
    "food_drink": "4.food_drink_model_uk_ie_lovable.pkl",  # missing in bucket -> will be skipped
    "meal_type": "5.meal_type_model_multilabel_lovable.pkl",
}

SCHEMA = "public"
TABLE = "static test"  # contains a space => quoted
PK_COL = "review_link"
TEXT_COL = "review_text"

BATCH_SIZE = 256
REQUEST_TIMEOUT = 60
WHERE_CLAUSE_RESUME = "sentiment_label IS NULL"

FOOD_SAFETY_IDX_TO_LABEL = {
    0: "Food Poisoning",
    1: "Poor Hygiene",
    2: "Allergens",
    3: "Food Safety issue not classified",
}

TOP_K = 3
MIN_P = 0.15

# =======================
# Env vars
# =======================
SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_STORAGE_URL = os.getenv("SUPABASE_STORAGE_URL")
SUPABASE_DB_URL = os.getenv("SUPABASE_DB_URL")
SUPABASE_SERVICE_ROLE_KEY = os.getenv("SUPABASE_SERVICE_ROLE_KEY")  # optional for public downloads, useful for DB

if not SUPABASE_URL or not SUPABASE_DB_URL:
    print("ERROR: missing env vars. Required: SUPABASE_URL, SUPABASE_DB_URL")
    sys.exit(1)

logging.basicConfig(level=logging.INFO, format="%(asctime)s %(levelname)s %(message)s")

# =======================
# Storage helpers (PUBLIC)
# =======================
def storage_base() -> str:
    return (SUPABASE_STORAGE_URL or (SUPABASE_URL.rstrip("/") + "/storage/v1")).rstrip("/")

def storage_public_url(bucket: str, key: str) -> str:
    return f"{storage_base()}/object/public/{bucket}/{key.lstrip('/')}"

def download_model_to_memory(bucket: str, key: str) -> bytes:
    url = storage_public_url(bucket, key)
    logging.info(f"Downloading model: {key}")
    resp = requests.get(url, timeout=REQUEST_TIMEOUT)
    if resp.status_code != 200:
        raise RuntimeError(f"Failed download {key}: {resp.status_code} {resp.text[:300]}")
    return resp.content

def load_model_from_bytes(b: bytes):
    return joblib.load(io.BytesIO(b))

# =======================
# DB helpers
# =======================
def connect_db(db_url: str):
    conn = psycopg2.connect(db_url)
    conn.autocommit = False
    return conn

def quoted_table_name(schema: str, table: str) -> str:
    return f'"{schema}"."{table}"'

def fetch_batch(conn, batch_size: int) -> List[Tuple[str, str]]:
    qt = quoted_table_name(SCHEMA, TABLE)
    sql_txt = f"""
      SELECT "{PK_COL}"::text, "{TEXT_COL}"::text
      FROM {qt}
      WHERE {WHERE_CLAUSE_RESUME}
      LIMIT %s;
    """
    with conn.cursor() as cur:
        cur.execute(sql_txt, (batch_size,))
        return cur.fetchall()

# =======================
# Model output helpers
# =======================
def predict_sentiment(model, texts: List[str]) -> List[str]:
    return [str(x) for x in model.predict(texts)]

def predict_food_safety_primary(model, texts: List[str]) -> Tuple[List[str], List[float]]:
    pred = np.asarray(model.predict(texts))
    proba = model.predict_proba(texts) if hasattr(model, "predict_proba") else None

    class_ids = list(model.named_steps["clf"].classes_)
    labels = [FOOD_SAFETY_IDX_TO_LABEL.get(int(c), str(c)) for c in class_ids]

    if proba is None:
        prim, prim_p = [], []
        for row in pred:
            idxs = np.where(row == 1)[0]
            if len(idxs) == 0:
                prim.append("Unclassified")
                prim_p.append(np.nan)
            else:
                prim.append(labels[int(idxs[0])])
                prim_p.append(np.nan)
        return prim, prim_p

    best_idx = np.argmax(proba, axis=1)
    prim = [labels[int(i)] for i in best_idx]
    prim_p = [float(proba[r, int(best_idx[r])]) for r in range(len(texts))]
    return prim, prim_p

def predict_multioutput_2cols(model, texts: List[str]) -> Tuple[List[str], List[str]]:
    arr = np.asarray(model.predict(texts))
    if arr.ndim == 1:
        return [str(x) for x in arr], [None for _ in arr]
    if arr.shape[1] < 2:
        return [str(x) for x in arr[:, 0]], [None for _ in arr[:, 0]]
    return [str(x) for x in arr[:, 0]], [str(x) for x in arr[:, 1]]

def mapped_label_names_ovr(model) -> List[str]:
    class_ids = list(model.named_steps["clf"].classes_)
    if len(class_ids) and isinstance(class_ids[0], str):
        return class_ids
    return [str(c) for c in class_ids]

def topk_multilabel(model, texts: List[str], k=3, min_p=0.15,
                   exclude=("Food/Drink not classified", "Food not classified", "Meal type not classified")):
    proba = model.predict_proba(texts)
    labels = mapped_label_names_ovr(model)
    exclude = set(exclude)

    out = []
    for row in proba:
        pairs = [(lab, float(p)) for lab, p in zip(labels, row) if lab not in exclude]
        pairs.sort(key=lambda x: x[1], reverse=True)
        pairs = [p for p in pairs if p[1] >= min_p][:k]
        out.append(pairs)
    return out

# =======================
# Bulk update
# =======================
def apply_updates(conn, updates: List[Dict[str, Any]]):
    if not updates:
        return

    qt = quoted_table_name(SCHEMA, TABLE)

    cols = [
        "pk",
        "sentiment_label",
        "food_safety_primary_label",
        "food_safety_primary_p",
        "service_label",
        "food_quality_label",
        "food_drink_1","food_drink_1_p","food_drink_2","food_drink_2_p","food_drink_3","food_drink_3_p",
        "meal_type_1","meal_type_1_p","meal_type_2","meal_type_2_p","meal_type_3","meal_type_3_p",
    ]

    values = [tuple(u.get(c) for c in cols) for u in updates]

    with conn.cursor() as cur:
        cur.execute("CREATE TEMP TABLE tmp_preds (pk text) ON COMMIT DROP;")
        for col in cols[1:]:
            if col.endswith("_p"):
                cur.execute(f'ALTER TABLE tmp_preds ADD COLUMN "{col}" double precision;')
            else:
                cur.execute(f'ALTER TABLE tmp_preds ADD COLUMN "{col}" text;')

        col_sql = sql.SQL(",").join(sql.Identifier(c) for c in cols)
        insert_sql = sql.SQL("INSERT INTO tmp_preds ({}) VALUES %s").format(col_sql)
        execute_values(cur, insert_sql, values)

        set_clause = ", ".join([f'"{c}" = data."{c}"' for c in cols[1:]])
        cur.execute(f"""
            UPDATE {qt} AS t
            SET {set_clause}
            FROM tmp_preds AS data
            WHERE t."{PK_COL}"::text = data.pk;
        """)

    conn.commit()

# =======================
# Main
# =======================
def main():
    # Download + load models, skip missing
    loaded: Dict[str, Any] = {}
    missing = []

    for name, key in MODEL_KEYS.items():
        try:
            b = download_model_to_memory(BUCKET_NAME, key)
            loaded[name] = load_model_from_bytes(b)
        except Exception as e:
            logging.warning(f"Skipping model '{name}' ({key}) due to download/load error: {e}")
            missing.append(name)

    if not loaded:
        logging.error("No models loaded. Check bucket contents.")
        sys.exit(1)

    logging.info(f"Models loaded: {sorted(list(loaded.keys()))}")
    if missing:
        logging.warning(f"Models missing: {missing}")

    conn = connect_db(SUPABASE_DB_URL)
    logging.info("Connected to database.")

    total = 0
    try:
        while True:
            rows = fetch_batch(conn, BATCH_SIZE)
            if not rows:
                logging.info("No more rows to process.")
                break

            pks = [r[0] for r in rows]
            texts = [(r[1] or "").strip() for r in rows]

            sent = [None] * len(rows)
            fs_primary = [None] * len(rows)
            fs_primary_p = [None] * len(rows)
            service_label = [None] * len(rows)
            food_quality_label = [None] * len(rows)
            fd_top = [[] for _ in rows]
            mt_top = [[] for _ in rows]

            if "sentiment" in loaded:
                sent = predict_sentiment(loaded["sentiment"], texts)

            if "food_safety" in loaded:
                fs_primary, fs_primary_p = predict_food_safety_primary(loaded["food_safety"], texts)

            if "svc_foodqual" in loaded:
                service_label, food_quality_label = predict_multioutput_2cols(loaded["svc_foodqual"], texts)

            if "food_drink" in loaded:
                fd_top = topk_multilabel(
                    loaded["food_drink"], texts, k=TOP_K, min_p=MIN_P,
                    exclude=("Food/Drink not classified","Food not classified")
                )

            if "meal_type" in loaded:
                mt_top = topk_multilabel(
                    loaded["meal_type"], texts, k=TOP_K, min_p=MIN_P,
                    exclude=("Meal type not classified",)
                )

            updates = []
            for i in range(len(rows)):
                u = {
                    "pk": pks[i],
                    "sentiment_label": sent[i],
                    "food_safety_primary_label": fs_primary[i],
                    "food_safety_primary_p": fs_primary_p[i],
                    "service_label": service_label[i],
                    "food_quality_label": food_quality_label[i],
                }

                for j in range(1, TOP_K + 1):
                    u[f"food_drink_{j}"] = fd_top[i][j-1][0] if len(fd_top[i]) >= j else None
                    u[f"food_drink_{j}_p"] = fd_top[i][j-1][1] if len(fd_top[i]) >= j else None

                for j in range(1, TOP_K + 1):
                    u[f"meal_type_{j}"] = mt_top[i][j-1][0] if len(mt_top[i]) >= j else None
                    u[f"meal_type_{j}_p"] = mt_top[i][j-1][1] if len(mt_top[i]) >= j else None

                updates.append(u)

            apply_updates(conn, updates)
            total += len(updates)
            logging.info(f"Processed {len(updates)} rows. Total processed: {total}")
            time.sleep(0.15)

    except Exception:
        logging.exception("Unhandled error")
        conn.rollback()
        raise
    finally:
        conn.close()
        logging.info("DB connection closed.")

if __name__ == "__main__":
    main()

```

```python
from supabase import create_client
import requests
from datetime import datetime, timezone

BASE = "https://api.content.tripadvisor.com/api/v1"
REVIEWS_PER_LOCATION = 5

def fetch_reviews(session, api_key, loc_id, limit):
    url = f"{BASE}/location/{loc_id}/reviews"
    params = {"key": api_key, "limit": limit}
    r = session.get(url, params=params, headers={"Accept": "application/json"}, timeout=30)
    r.raise_for_status()
    return r.json()

def normalize_review(loc_id, review):
    rid = review.get("id")
    if rid is None:
        return None
    return {
        "tripadvisor_review_id": str(rid),
        "tripadvisor_location_id": str(loc_id),
        "published_date": review.get("published_date"),
        "rating": review.get("rating"),
        "lang": review.get("lang"),
        "title": review.get("title"),
        "text": review.get("text"),
        "url": review.get("url"),
        "fetched_at": datetime.now(timezone.utc).isoformat(),
    }

def upsert_reviews(sb, loc_id, reviews):
    rows = []
    for rev in reviews:
        row = normalize_review(loc_id, rev)
        if row:
            rows.append(row)
    if not rows:
        return 0
    sb.table("ta_reviews").upsert(rows, on_conflict="tripadvisor_review_id").execute()
    return len(rows)

def mark_fetched(sb, loc_id):
    sb.table("ta_locations").update({"last_fetched_at": "now()", "updated_at": "now()"}).eq(
        "tripadvisor_location_id", loc_id
    ).execute()

sb = create_client(os.environ["SUPABASE_URL"], os.environ["SUPABASE_SERVICE_ROLE_KEY"])
session = requests.Session()

# Pick ONE location id that you know should have reviews:
LOC_ID = "10025788"

payload = fetch_reviews(session, os.environ["TRIPADVISOR_API_KEY"], LOC_ID, REVIEWS_PER_LOCATION)
print("Payload keys:", payload.keys())
reviews = payload.get("data", [])
print("Reviews returned:", len(reviews))

n = upsert_reviews(sb, LOC_ID, reviews)
mark_fetched(sb, LOC_ID)
print("Upserted:", n)
```
